### –ó–∞–¥–∞—á–∞ 1 [1 –±–∞–ª–ª].
–ë—É—Ç–µ—Ä–±—Ä–æ–¥ –ø–æ–¥–±—Ä–æ—Å–∏–ª–∏ 200 —Ä–∞–∑, 105 —Ä–∞–∑ –æ–Ω —É–ø–∞–ª –º–∞—Å–ª–æ–º –≤–Ω–∏–∑ –∏ 95 –º–∞—Å–ª–æ–º –≤–≤–µ—Ä—Ö. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ –±—É—Ç–µ—Ä–±—Ä–æ–¥ —Å —Ä–∞–≤–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –ø–∞–¥–∞–µ—Ç –º–∞—Å–ª–æ–º –≤–≤–µ—Ä—Ö –∏ –≤–Ω–∏–∑.

–ü—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–æ—Å—Ç—É—é –≥–∏–ø–æ—Ç–µ–∑—É:
$$ H_0: p=0.5 $$
$$ H_a: p!=0.5 $$

$$ X_1, ... , X_n ‚âÉ iid \space Bern(p) $$

–ò—Å–ø–æ–ª—å–∑—É–µ–º Z-—Ç–µ—Å—Ç(–¶–ü–¢)
–û—Ü–µ–Ω–∏–≤–∞–µ–º p –ø–æ –≤—ã–±–æ—Ä–∫–µ:

$$  \widehat{p} = \frac{105}{200} = 0.525 $$

–ü–æ –¶–ü–¢:

$$ \widehat{p} ‚âÉ N(0.5, \frac{0.5(1-0.5)}{n}) $$

$$ z ‚âÉ \frac{\widehat{p} - 0.5}{\sqrt{\frac{0.5(1-0.5)}{n}}}  ‚âÉ N(0,1) $$

$$ z_{obs} = 0.707 $$

–ï—Å–ª–∏ –≤–æ–∑—å–º–µ–º Œ±=0.05 , —Ç–æ
$$  z_{obs} ‚àà (-1.96,1.96) $$
–ó–Ω–∞—á–∏—Ç –≥–∏–ø–æ—Ç–µ–∑—É –ù–ï –û–¢–í–ï–†–ì–ê–ï–ú

```python
n = 200
verh = 105
p_obs = verh/n
z = (p_obs - 0.5) / (0.5*(1-0.5)/n)**0.5
```

```python
from statsmodels.stats.proportion import proportions_ztest

verh = 105
n = 200
value = 0.5

z_stat, p_value = proportions_ztest(verh, n, value, alternative='two-sided')

print(f"Z-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {z_stat:.4f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("–û–¢–í–ï–†–ì–ê–ï–ú")
else:
    print("–ù–ï –û–¢–í–ï–†–ì–ê–ï–ú")
```

### –ó–∞–¥–∞—á–∞ 2 [2 –±–∞–ª–ª–∞].
–í –ó–∞–∫–æ–Ω–æ–¥–∞—Ç–µ–ª—å–Ω–æ–º —Å–æ–±—Ä–∞–Ω–∏–∏ –Ω–µ–∫–æ—Ç–æ—Ä–æ–π —Å—Ç—Ä–∞–Ω—ã (450 –¥–µ–ø—É—Ç–∞—Ç–æ–≤) –æ–±—Å—É–∂–¥–∞—é—Ç –≤–≤–µ–¥–µ–Ω–∏–µ –Ω–∞–ª–æ–≥–∞ –Ω–∞ –≤–æ–∑–¥—É—Ö. –ê–≥–µ–Ω—Ç –§–°–ë, –ø–æ–¥—Å–ª—É—à–∞–≤—à–∏–π –∑–∞—Å–µ–¥–∞–Ω–∏–µ, —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ –ó–ê –ø–ª–∞–Ω–∏—Ä—É–µ—Ç –ø—Ä–æ–≥–æ–ª–æ—Å–æ–≤–∞—Ç—å –æ–∫–æ–ª–æ 70% –¥–µ–ø—É—Ç–∞—Ç–æ–≤. –ó–∞–∫–æ–Ω –±—É–¥–µ—Ç –ø—Ä–∏–Ω—è—Ç, –µ—Å–ª–∏ –ó–ê –ø—Ä–æ–≥–æ–ª–æ—Å—É–µ—Ç –±–æ–ª–µ–µ –ø–æ–ª–æ–≤–∏–Ω—ã –¥–µ–ø—É—Ç–∞—Ç–æ–≤. –ö–∞–∫–æ–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ–ø—É—Ç–∞—Ç–æ–≤ –Ω—É–∂–Ω–æ –æ–ø—Ä–æ—Å–∏—Ç—å –Ω–∞ –≤—ã—Ö–æ–¥–µ —Å –∑–∞—Å–µ–¥–∞–Ω–∏—è, —á—Ç–æ–±—ã –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏–π, –ø—Ä–æ–≤–µ—Ä—è—é—â–∏–π –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ –∑–∞–∫–æ–Ω –Ω–µ –ø—Ä–∏–Ω—è—Ç, –ø—Ä–æ—Ç–∏–≤ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã (–∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ—Ç –∞–≥–µ–Ω—Ç–∞), —Å —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 5% –∏ –º–æ—â–Ω–æ—Å—Ç—å—é –Ω–µ –º–µ–Ω–µ–µ 80%?

–ü–æ–ª—É—á–∞–µ—Ç—Å—è —á—Ç–æ –¥–µ–ø—É—Ç–∞—Ç—ã –∑–∞–∫—Å–æ–±—Ä–∞–Ω–∏—è - –Ω–∞—à–∞ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–∞—è —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å.  
–ù–∞—à–∞ –∑–∞–¥–∞—á–∞ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π n.  
–ù–∞—à–∞ –≥–∏–ø–æ—Ç–µ–∑–∞:
$$ H_0: p \leq 0.5 $$
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è:
$$ H_a: p = 0.7 $$

$$ X_1, ... , X_n ‚âÉ iid \space Bern(p) $$
–ü–æ —É—Å–ª–æ–≤–∏—è –Ω–∞–º –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π n, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ:
1. Œ±=0.05
2. 1 - Œ≤ ‚â• 0.8
–¢–æ –µ—Å—Ç—å
$$ Œ± = P(H_0 –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ | H_0 –≤–µ—Ä–Ω–∞) $$
$$ Œ≤ = P(H_0 –Ω–µ –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ | H_0 –Ω–µ –≤–µ—Ä–Ω–∞) $$
–í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏–º Z-—Ç–µ—Å—Ç–æ–º –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∫—Ä–∏—Ç–µ—Ä–∏—è.
$$ Z_{obs0} = \frac{\widehat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}} $$
–ò—Å–ø–æ–ª—å–∑—É–µ–º –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π.
–¢–æ–≥–¥–∞:
$$ Œ± = P(H_0 –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ | H_0 –≤–µ—Ä–Ω–∞) = P(Z_{obs0} > Z_{1 - Œ±} | p \leq 0.5) $$
$$ Œ≤ = P(H_0 –Ω–µ –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ | H_0 –Ω–µ –≤–µ—Ä–Ω–∞) = P(Z_{obs0} \leq Z_{1 - Œ±} | p = 0.7)  $$
–†–∞—Å–ø–∏—à–µ–º  Œ≤
$$ P(Z_{obs0} \leq Z_{1 - Œ±} | p = 0.7) = P(\frac{\widehat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}}  \leq Z_{1 - Œ±} | p = 0.7) $$
–ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —á—Ç–æ–±—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —à–ª–æ —Å
$$ Z_{obs1} = \frac{\widehat{p} - p_a}{\sqrt{\frac{p_a(1-p_a)}{n}}} $$
$$ P(\widehat{p}  \leq Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{n}} + p_0 | p = 0.7)
  = $$
 $$ P( \frac{\widehat{p} - p_a}{\sqrt{\frac{p_a(1-p_a)}{n}}} \leq \frac{Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{n}} + p_0  - p_a} {\sqrt{\frac{p_a(1-p_a)}{n}}}| p = 0.7)
 $$

–ù–∞—à–∞ –∑–∞–¥–∞—á–∞ –Ω–∞–π—Ç–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π n,–≤–æ–∑–≤—Ä–∞—â–∞–µ–º—Å—è –∫ —É—Ä–∞–≤–Ω–µ–Ω–∏—é
$$ Œ≤ = P(H_0 –Ω–µ –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç–∞ | H_0 –Ω–µ –≤–µ—Ä–Ω–∞)  $$
–ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏
$$ 0.2 \leq P( \frac{\widehat{p} - p_a}{\sqrt{\frac{p_a(1-p_a)}{n}}} \leq \frac{Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{n}} - p_0  - p_a} {\sqrt{\frac{p_a(1-p_a)}{n}}})  $$
–ß—É—Ç—å –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –ª–µ–≤—É—é —á–∞—Å—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏—è
$$ 0.2 = P(X < Z_{0.2}) = ùû•(Z_{0.2}) $$ –≥–¥–µ ùû•(x) - —Ñ—É–Ω–∫—Ü–∏—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–≥–Ω–∏—è
–ß—É—Ç—å –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –ø—Ä–∞–≤—É—é —á–∞—Å—Ç—å
$$ P( \frac{\widehat{p} - p_a}{\sqrt{\frac{p_a(1-p_a)}{n}}} \leq \frac{Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{n}} + p_0  - p_a} {\sqrt{\frac{p_a(1-p_a)}{n}}}) =ùû•(\frac{Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{n}} } {\sqrt{\frac{p_a(1-p_a)}{n}}} + \frac{p_0  - p_a} {\sqrt{\frac{p_a(1-p_a)}{n}}}) =  ùû•(Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{p_a(1-p_a)}} + \frac{p_0  - p_a} {\sqrt{\frac{p_a(1-p_a)}{n}}})$$
–¢–µ–ø–µ—Ä—å –∏—Å—Ö–æ–¥–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤—ã–≥–ª—è–¥–∏—Ç —Ç–∞–∫

$$ ùû•(Z_{Œ≤}) = ùû•(Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{p_a(1-p_a)}} + \frac{p_0  - p_a} {\sqrt{\frac{p_a(1-p_a)}{n}}}) $$
$$ Z_{Œ≤} = Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{p_a(1-p_a)}} + \frac{p_0  - p_a} {\sqrt{\frac{p_a(1-p_a)}{n}}} $$
$$ \frac{(Z_{Œ≤} - Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{p_a(1-p_a)}}) (p_a(1-p_a))}{p_0  - p_a} = \sqrt{n} $$
$$ n = \frac{(-Z_{1-Œ≤} - Z_{1 - Œ±} * \sqrt{\frac{p_0(1-p_0)}{p_a(1-p_a)}}) (\sqrt{p_a(1-p_a)})}{p_0  - p_a}^2 $$
$$ n = (\frac{-Z_{1-Œ≤}(\sqrt{p_a(1-p_a)}) - Z_{1 - Œ±}\sqrt{p_0(1-p_0)} }{p_0-p_a})^2 $$

```python
Z_b = 0.85
Z_a = 1.65
p_0 = 0.5
p_a = 0.7
n = ((-Z_b * (p_a*(1-p_a))**0.5 - Z_a * (p_0*(1-p_0))**0.5)/(p_0 - p_a))**2
print(n)
```

### –ó–∞–¥–∞—á–∞ 3 [3 –±–∞–ª–ª–∞].
–í—Ä–µ–º—è –≤ —á–∞—Å–∞—Ö –±–µ–∑–æ—Ç–∫–∞–∑–Ω–æ–π —Ä–∞–±–æ—Ç—ã –º–∏–∫—Ä–æ–Ω–∞—É—à–Ω–∏–∫–∞, –≤–µ–ª–∏—á–∏–Ω–∞ $x$, –ø–æ–¥—á–∏–Ω—è–µ—Ç—Å—è —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–º—É
(–ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–º—É) –∑–∞–∫–æ–Ω—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º $\lambda$:

\begin{equation*}
f(x, \lambda) =
 \begin{cases}
   \lambda e^{-\lambda x}, x \ge 0\\
   0, x < 0
 \end{cases}
\end{equation*}

–ü–æ –≤—ã–±–æ—Ä–∫–µ –∏–∑ 100 –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –Ω–∞–±–ª—é–¥–µ–Ω–∏–π $\overline{x}$ = 1.

–° –ø–æ–º–æ—â—å—é –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏—Ö —Å–≤–æ–π—Å—Ç–≤ –æ—Ü–µ–Ω–æ–∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è –ø–æ—Å—Ç—Ä–æ–π—Ç–µ 95%-—ã–π –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:

–∞) –î–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\lambda$

–±) –î–ª—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ, —á—Ç–æ –Ω–∞—É—à–Ω–∏–∫ –ø—Ä–æ—Ä–∞–±–æ—Ç–∞–µ—Ç –±–µ–∑ —Å–±–æ–µ–≤ –≤—Å—é –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—É—é ‚Äî 80 –º–∏–Ω—É—Ç

# –ü—É–Ω–∫—Ç –ê.  
–ü–æ—Å—Ç—Ä–æ–∏–º –û–ú–ü –¥–ª—è Œª
$$ \prod Œªe^{-Œªx} = Œª^ne^{-Œª\sum x_i} $$
–õ–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä—É–µ–º
$$ ln(Œª^ne^{-Œª\sum x_i} ) = ln( n ln(Œª) - Œª\sum x_i) $$
–ü—Ä–æ–¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏—Ä—É–µ–º
$$ \frac{d(ln( n ln(Œª) - Œª\sum x_i))}{dŒª} = \frac{n}{\lambda} - \sum x_i = \frac{n}{\lambda} - n\widehat{x}  $$
–ü—Ä–∏–≤—Ä–∞–≤–Ω–∏–≤–∞–µ–º –∫ –Ω—É–ª–±
$$  \frac{n}{\lambda} - n\widehat{x}   = 0 $$
$$ Œª = \frac{1}{\widehat{x}} = 1 $$
–¢–æ–≥–¥–∞ –º–∞—Ç –æ–∂–∏–¥–∞–Ω–∏–µ
$$ Œº = Œª^{-1} = 1$$
–î–ª—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –Ω–∞—Ö–æ–¥–∏–º –≤—Ç–æ—Ä—É—é –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –æ—Ç –û–ú–ü
$$ \frac{d^2(ln( n ln(Œª) - Œª\sum x_i))}{dŒª^2} = - \frac{n}{\lambda^2}   $$
$$ œÉ = \frac{Œª^2}{n} = \frac{1}{100} $$

–¢–æ–≥–¥–∞ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –±—É–¥–µ—Ç:
$$ (\widehat{Œª} - Z_{\frac{Œ±}{2}}\sqrt{\frac{\widehat{Œª}^2 }{100}}, \widehat{Œª} + Z_{\frac{Œ±}{2}}\sqrt{\frac{\widehat{Œª}^2}{100}}) $$

–ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º
$$ (0.80,1.19) $$

# –ü—É–Ω–∫—Ç –ë.  
–¢—Ä–µ–±—É–µ—Ç—Å—è –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è:
$$ P(X \geq 1.3) = e^{-1.3Œª} $$
–í–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –¥–µ–ª—å—Ç–∞-–º–µ—Ç–æ–¥–æ–º
$$ g(\widehat{Œª}) =  P(X \geq 1.3) =  e^{-1.3\widehat{Œª}} $$
$$ g(\widehat{Œª}) \sim N(g(\lambda), \widehat{œÉ}g'(\widehat{Œª})^2 ) $$
–ù–∞–π–¥–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é
$$ g'(\widehat{Œª}) = -1.3e^{-1.3\widehat{Œª}} $$
–ò—Ç–æ–≥:
$$  P(X \geq 1.3) \sim N(e^{-1.3}, \frac{1}{100}(-1.3e^{-1.3})^2 ) $$
–¢–æ–≥–¥–∞ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:
$$ (e^{-1.3} - Z_{\frac{Œ±}{2}}\sqrt{\frac{1}{100}(-1.3e^{-1.3})^2}, e^{-1.3} + Z_{\frac{Œ±}{2}}\sqrt{\frac{1}{100}(-1.3e^{-1.3})^2}) $$
–ü–æ–¥—Å—á–∏—Ç–∞–µ–º:
(0.195, 0.332)

```python
import numpy as np
import scipy.stats as stats

n = 100
x_bar = 1
lambda_hat = 1 / x_bar
alpha = 0.05
z = stats.norm.ppf(1 - alpha/2)
se_lambda = np.sqrt(lambda_hat**2 / n)
ci_lambda = (lambda_hat - z * se_lambda, lambda_hat + z * se_lambda)

t = 80 / 60
p_hat = np.exp(-lambda_hat * t)
dg_dlambda = -t * np.exp(-lambda_hat * t)
var_p = (dg_dlambda**2) * (lambda_hat**2 / n)
se_p = np.sqrt(var_p)
ci_p = (p_hat - z * se_p, p_hat + z * se_p)

print(f"–∞) 95% –î–ò –¥–ª—è Œª: ({ci_lambda[0]:.3f}, {ci_lambda[1]:.3f})")
print(f"–±) 95% –î–ò –¥–ª—è P(X > 80 –º–∏–Ω): ({ci_p[0]:.3f}, {ci_p[1]:.3f})")
```

### –ó–∞–¥–∞—á–∞ 4 [1 –±–∞–ª–ª].
–ë–æ–≥–∞—Ç—ã—Ä—å –†–∞–¥–∏—Å–ª–∞–≤ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç, —á—Ç–æ –≤—Å–µ –≤–∏–¥—ã —è—â–µ—Ä–æ–≤ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–Ω–æ.
–û—Ç–ø—Ä–∞–≤–∏–≤—à–∏—Å—å –Ω–∞ –æ—Ö–æ—Ç—É –≤ —Ä–∞–π–æ–Ω –æ–∑–µ—Ä–∞ –ë–∞–π–∫–∞–ª, –†–∞–¥–∏—Å–ª–∞–≤ –ø–æ–π–º–∞–ª –¥–≤—É—Ö –≥–∏–ø–Ω–æ—è—â–µ—Ä–æ–≤,
–æ–¥–Ω–æ–≥–æ –¥–æ–ª–±–æ—è—â–µ—Ä–∞ –∏ –æ–¥–Ω–æ–≥–æ —è—â–µ—Ä–∞-–≤—Ç–∞—â–µ—Ä–∞.
–ü–æ–º–æ–≥–∏—Ç–µ –†–∞–¥–∏—Å–ª–∞–≤—É –æ—Ü–µ–Ω–∏—Ç—å –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–¥–æ–≤ —è—â–µ—Ä–æ–≤ —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞
–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è.

–£ –Ω–∞—Å –µ—Å—Ç—å –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–¥–æ–≤ —è—â–µ—Ä–æ–≤ - m.  
–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º —á—Ç–æ —ç—Ç—É —à—Ç—É–∫—É –æ–ø–∏—Å—ã–≤–∞–µ–º –º—É–ª—å—Ç–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ    
–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å:
$$ P(X_1 = n_1, ...,X_m = n_m) = \frac{n!}{n_1! * ... * n_m!}p_1^{n_1}* ... * p_m^{n_m} $$
–¢–∞–∫ –∫–∞–∫ –†–∞–¥–∏—Å–ª–∞–≤ —Å—á–∏—Ç–∞–µ—Ç —á—Ç–æ –≤—Å–µ —è—â–µ—Ä—ã —Ä–∞–≤–Ω–æ–≤–µ—Ä–æ—è—Ç–Ω—ã, —Ç–æ
$$ p_i = \frac{1}{m} $$
–ò —Ç–æ–≥–¥–∞
$$ P(X_1 = n_1, ...,X_m = n_m) = \frac{n!}{n_1! * ... * n_m!}(\frac{1}{m})^n $$
–ü–æ—Å—á–∏—Ç–∞–µ–º –ú–ú–ü, –≥–¥–µ –±—É–¥–µ–º –æ—Ü–µ–Ω–∏–≤–∞—Ç—å m:
$$ L(m) = \frac{n!}{n_1! * ... * n_m!}(\frac{1}{m})^n = \frac{4!}{2!} (\frac{1}{m})^2(\frac{m-1}{m})(\frac{m-2}{m}) = \frac{4!}{2!} \frac{(m-1)(m-2)}{m^4}$$
–õ–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä—É–µ–º
$$ ln(L(m)) = ln( \frac{4!}{2!} \frac{(m-1)(m-2)}{m^4}) = ln(\frac{4!}{2!}) + ln(m-1) + ln(m-2) - ln(m^4) $$
–î–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏—Ä—É–µ–º
$$ \frac{ ln(L(m))}{dm} = \frac{1}{m-1} + \frac{1}{m-2} -  4\frac{1}{m} = 0$$
–ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º
$$2m^2-9m+8=0$$
–¢–∞–∫ –∫–∞–∫ —Ç–æ—á–∫–∏, –Ω–æ –≤ –ø–µ—Ä–≤–æ–π m = 1.2 - —á–µ–≥–æ –±—ã—Ç—å –Ω–µ –º–æ–∂–µ—Ç
–ø–æ—ç—Ç–æ–º—É –≤—ã–±–∏—Ä–∞–µ–º –≤—Ç–æ—Ä—É—é —Ç–æ—á–∫—É m = 3.2.  
–ò –æ—Ç–≤–µ—Ç: 3

```python
def L(k):
    return (k-1)*(k-2) / (k**4)
res = []
for m in range(3, 10):
    print(f"m={m}: L(m)={L(m):.4f}")
    res.append((L(m), m))
res.sort()
print(f"–û—Ç–≤–µ—Ç: {res[-1][1]}")
```

### –ó–∞–¥–∞—á–∞ 5 [3 –±–∞–ª–ª–∞].

–ì–ª–µ–± –ñ–µ–≥–ª–æ–≤ —Ä–µ—à–∞–µ—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, –Ω–µ –∂—É–ª—å–Ω–∏—á–∞–µ—Ç –ª–∏ –ö–æ–ø—á—ë–Ω—ã–π –ø—Ä–∏ –∏–≥—Ä–µ –≤ –Ω–∞–ø–µ—Ä—Å—Ç–∫–∏, –¥–ª—è —á–µ–≥–æ –ø—Ä–µ–¥–ª–∞–≥–∞–µ—Ç —Ç–æ–º—É —Å—ã–≥—Ä–∞—Ç—å 5 –ø–∞—Ä—Ç–∏–π. –ñ–µ–≥–ª–æ–≤ —Ä–µ—à–∞–µ—Ç, —á—Ç–æ –≤ –∫–∞–∂–¥–æ–π –ø–∞—Ä—Ç–∏–∏ –±—É–¥–µ—Ç –≤—ã–±–∏—Ä–∞—Ç—å –æ–¥–∏–Ω –∏–∑ —Ç—Ä—ë—Ö –Ω–∞–ø—ë—Ä—Å—Ç–∫–æ–≤ –Ω–∞—É–≥–∞–¥, –Ω–µ —Å–º–æ—Ç—Ä—è –Ω–∞ –¥–≤–∏–∂–µ–Ω–∏—è —Ä—É–∫ –≤–µ–¥—É—â–µ–≥–æ. –û—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –ö–æ–ø—á—ë–Ω—ã–π —á–µ—Å—Ç–µ–Ω, –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ —É–≥–∞–¥–∞—Ç—å –Ω–∞–ø—ë—Ä—Å—Ç–æ–∫, –ø–æ–¥ –∫–æ—Ç–æ—Ä—ã–º —Å–ø—Ä—è—Ç–∞–Ω —à–∞—Ä–∏–∫, —Ä–∞–≤–Ω–∞ 1/3. –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –ö–æ–ø—á—ë–Ω—ã–π –∫–∞–∫–∏–º-—Ç–æ –æ–±—Ä–∞–∑–æ–º –∂—É–ª—å–Ω–∏—á–∞–µ—Ç (–Ω–∞–ø—Ä–∏–º–µ—Ä, –Ω–µ–∑–∞–º–µ—Ç–Ω–æ –ø—Ä—è—á–µ—Ç —à–∞—Ä–∏–∫), —Ç–∞–∫ —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É–≥–∞–¥–∞—Ç—å –Ω—É–∂–Ω—ã–π –Ω–∞–ø—ë—Ä—Å—Ç–æ–∫ —Ä–∞–≤–Ω–∞ 1/5. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π: –æ—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è, –µ—Å–ª–∏ –¥–µ—Ç–µ–∫—Ç–∏–≤ –Ω–∏ —Ä–∞–∑—É –Ω–µ —É–≥–∞–¥–∞–µ—Ç, –≥–¥–µ —à–∞—Ä–∏–∫.

–∞) –ù–∞–π–¥–∏—Ç–µ —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è.

–±) –ù–∞–π–¥–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—à–∏–±–∫–∏ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–¥–∞.


$$ H_0: p = \frac{1}{3}$$
$$ H_a: p = \frac{1}{5}$$

# –ü—É–Ω–∫—Ç –ê.
–ù–∞–º —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ Œ±:
$$ Œ± = P(H_0 \space –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å| H_0 \space –≤–µ—Ä–Ω–∞)$$
–í —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –Ω–∞–º –Ω–∞–¥–æ –Ω–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ñ–µ–≥–ª–æ–≤ –Ω–∏ —Ä–∞–∑—É –Ω–µ –≤—ã–∏–≥—Ä–∞–µ—Ç –ø—Ä–∏ p = \frac{1}{3}.   
$$ (1-p) ^ n = \frac{2}{3}^5 = \frac{32}{243} = 0.13$$
# –ü—É–Ω–∫—Ç –ë.
–ù–∞–º —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ Œ±:
$$ Œ≤ = P(H_0 \space –Ω–µ \space –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å| H_0 \space –Ω–µ \space –≤–µ—Ä–Ω–∞)$$
–í —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –Ω–∞–º –Ω–∞–¥–æ –Ω–∞–π—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ñ–µ–≥–ª–æ–≤ —Ö–æ—Ç—å —Ä–∞–∑ –≤—ã–∏–≥—Ä–∞–µ—Ç –ø—Ä–∏ p = \frac{1}{5}.   
$$ 1 - (1-p) ^ n = 1 - \frac{4}{5}^5 = 0.67$$
